# plot a horizontal bar chart to analyze the reduced days of activity due to illness based on Gender

```python
db= df.groupby('gender')['reduced'].sum().to_frame().reset_index()
plt.barh(db['gender'],db['reduced'],color = ['cornflowerblue','darkred'])
plt.title('Bar Chart')
plt.xlabel('gender')
plt.ylabel('reduced activity')
plt.show()
```

# Visualize the percentage of people getting govt health Insurance due to low income, due to old age and also the percentge of people having private health insurance

```python
label=['Yes','No']
Y = df[df['freepoor']=='yes']
N = df[df['freepoor']=='no']
x = [Y.shape[0],N.shape[0]]
plt.figure(figsize=(3,3))
plt.pie(x,labels=label)
plt.title("% of people getting government health insurance due to low income")
plt.show()
 
Y = df[df['private']=='yes']
N = df[df['private']=='no']
x = [Y.shape[0],N.shape[0]]
plt.figure(figsize=(3,3))
plt.pie(x,labels=label)
plt.title("% of people having private health insurance")
plt.show()
 
Y = df[df['freerepat']=='yes']
N = df[df['freerepat']=='no']
x = [Y.shape[0],N.shape[0]]
plt.figure(figsize=(3,3))
plt.pie(x,labels=label)
plt.title("% of people getting government health insurance due to oldage,disability or veteran status")
plt.show()
```
